```python
import numpy as np
import pandas as pd

from sklearn import tree
from sklearn import metrics
from sklearn.cluster import KMeans
from sklearn.naive_bayes import BernoulliNB
from sklearn.metrics import classification_report
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
```

```python
# Leitura do arquivo csv
df_train = pd.read_csv('./data/train.csv')
df_test = pd.read_csv('./data/test.csv')

# Troca de NaN para 0
df_train = df_train.replace(np.nan, 0)
df_test = df_test.replace(np.nan, 0)

# Trocar dados na coluna 'Sex' para numéricos
df_train = df_train.replace(['male'], 0)
df_train = df_train.replace(['female'], 1)
df_test = df_test.replace(['male'], 0)
df_test = df_test.replace(['female'], 1)

# Salvar o PassengerId
test_passenger_id = df_test['PassengerId']
```

```python
df_train.head()
```

```python
df_test.head()
```

```python
# Remover a coluna de Survived
useful_columns = ['PassengerId', 'Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare']
df_train_without_survived = df_train[useful_columns]

# Dataframe com apenas a coluna Survived
df_survived = df_train['Survived']

# Treino do algoritmo
X_train, X_test, Y_train, Y_test = train_test_split(df_train_without_survived, df_survived, test_size=0.2, random_state=0)
```

<h2> Rede Neural </h2>

```python
import seaborn as sns
import tensorflow as tf
import matplotlib.pyplot as plt
from sklearn.preprocessing import Binarizer
```

```python
X_survived = df_train_without_survived
Y_survived = df_survived
```

```python
## 1. Instanciar o modelo
model = tf.keras.models.Sequential(name = 'Sequential_Model')

## 2. Adicionar as camadas
model.add(tf.keras.layers.Dense(units = 8,
                                activation = 'relu',
                                input_shape = (X_survived.shape[1], ),
                                name = 'input'))
model.add(tf.keras.layers.Dense(units = 1,
                                activation = 'linear',
                                name = 'output'))

## 3. Compilar
model.compile(optimizer = 'adam', loss = 'mse')
print(model.summary())

## 4. Fit
model.fit(X_survived, Y_survived, epochs = 100, batch_size = 1, validation_split=0.2, verbose = 0)
```

```python
# Criar as predições
yhat_survived = model.predict(df_test[useful_columns])[:,0]
```

```python
# Transformando em binário as predições obtidas
binarizer = Binarizer(0.5)
test_predict_result = binarizer.fit_transform(yhat_survived.reshape(-1, 1))
test_predict_result = test_predict_result.astype(np.int32)
```

```python
passenger_id = test_passenger_id.copy()
evaluation = passenger_id.to_frame()
evaluation["Survived"] = test_predict_result

evaluation[:10]
```

```python
evaluation.to_csv('./data/result.csv', index=False)
```

<h2> Auto-ML </h2>

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.feature_extraction import FeatureHasher
from scipy.stats import uniform, randint

import autosklearn.classification
```

```python
# ensemble_size: tamanho do end_ensemble (mínimo de 1)
# time_left_for_this_task: número de segundos que a task vai executar
# per_run_time_limit: número máximo de segundos alocados por modelo
model = autosklearn.classification.AutoSklearnClassifier(ensemble_size=10, 
                                                         time_left_for_this_task=120, 
                                                         per_run_time_limit=30)

# Começa o fitting do modelo de busca
model.fit(X_train, Y_train)

# Mostra as estatísticas
print(model.sprint_statistics())

# Recebe as predições do modelo
y_predictions = model.predict(df_test[useful_columns])
```

```python
passenger_id = test_passenger_id.copy()
evaluation = passenger_id.to_frame()
evaluation["Survived"] = y_predictions

evaluation[:10]
```

```python
evaluation.to_csv('./data/result-automl.csv', index=False)
```
